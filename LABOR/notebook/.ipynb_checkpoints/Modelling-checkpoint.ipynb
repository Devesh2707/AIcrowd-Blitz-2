```python
df['1/2'] = df[num_cols[0]]/df[num_cols[1]]
df['2/3'] = df[num_cols[1]]/df[num_cols[2]]
df['2-1'] = df[num_cols[1]] - df[num_cols[0]]
df['3-2'] = df[num_cols[2]] - df[num_cols[1]]
df['1/w'] = df[num_cols[0]]/df[num_cols[3]]
df['2/w'] = df[num_cols[1]]/df[num_cols[3]]
df['3/w'] = df[num_cols[2]]/df[num_cols[3]]
df['2-1/w'] = df['2-1']/df[num_cols[3]]
df['3-1/w'] = df['3-2']/df[num_cols[3]]
df['wxs'] = df[num_cols[3]] * df[num_cols[4]]
df['w/s'] = df[num_cols[3]]/df[num_cols[4]]
```
